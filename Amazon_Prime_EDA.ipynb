**Overview**

This project presents a detailed Exploratory Data Analysis (EDA) on a dataset containing content metadata from Amazon Prime Video, including movies and TV shows. Filling the null values in description with "No description available"

````python
#  description – Only 119 missing
df['description'] = df['description'].fillna("No description available")
````

There were 6484 missing values in age_certification columns , We will fill these empty values with "Unkown"

````python
# age_certification – 6484 missing (categorical)
df['age_certification'] = df['age_certification'].fillna("Unknown")
````

The seasons column has 8511 missing values , so replaceing the empty values with 0.

````python
# seasons – 8511 missing (mostly for MOVIE type)
# Set seasons to 0 for movies, and leave as NaN for shows if needed
df.loc[df['type'] == 'MOVIE', 'seasons'] = df.loc[df['type'] == 'MOVIE', 'seasons'].fillna(0)
````

The imdb_score column has 1021 missing values , we are filling the emplty values with the mean value.

````python
# imdb_score – 1021 missing (important for rating analysis)
df['imdb_score'] = df['imdb_score'].fillna(df['imdb_score'].mean())
````

The imdb_votes column has 1031 missing values , these we will fill by using votes value equal to median.

````python
# imdb_votes – 1031 missing (median)
df['imdb_votes'] = df['imdb_votes'].fillna(df['imdb_votes'].median())
````

The tmdb_popularity column has 547 null values , we will fill these with 0.

````python
# tmdb_popularity – 547 missing (can be filled with 0)
df['tmdb_popularity'] = df['tmdb_popularity'].fillna(0)
````

The tmdb_score has 2080 missing values , these can be filled with mean() values of the same column.

````python
# tmdb_score – 2080 missing (also important for quality analysis)
df['tmdb_score'] = df['tmdb_score'].fillna(df['tmdb_score'].mean())
````

The imdb_id has 667 columns with null values ,since these cannot be filled with any values, beacuse id are unique ,so were are dropping the rows having null values.

````python
df = df.drop(columns=['imdb_id'])
````

#### Chart - 1  **Distribution of content type**

````python
# Chart - 1 visualization code

# Create count plot
plt.figure(figsize=(4, 5))
ax = sns.countplot(data=df, x='type',color='green')
plt.title('Distribution of Content Type')

# Add value labels
for i in ax.patches:
    count = int(i.get_height())
    ax.annotate(f'{count}',
                (i.get_x() + i.get_width() / 2., i.get_height()),
                ha='center', va='bottom', fontsize=11, color='black')

plt.show()
````

##### 1. Why did you pick the specific chart?

We use countplot because it's ideal for visualizing the frequency of categorical variables. It automatically counts and displays the number of entries in each category, making it easy to compare values like Movies vs Shows in a clear, visual format.
