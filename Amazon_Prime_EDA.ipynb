# **Project Summary -**

**Overview**

This project presents a detailed Exploratory Data Analysis (EDA) on a dataset containing content metadata from Amazon Prime Video, including movies and TV shows. The aim is to understand patterns, trends, and insights related to content type, genre distribution, audience certification, release trends, ratings, runtime, and production countries.

By using Python's Pandas, Matplotlib, and Seaborn, the analysis focuses on extracting meaningful information that can benefit stakeholders such as content creators, data analysts, or business decision-makers in the OTT space.

**Dataset Features**

The dataset includes the following key attributes:

**id:** Unique identifier for each title

**title:** Name of the content

**type:** Movie or TV Show

**description:** Short summary of the content

**release_year:** Year of release

**age_certification:** Viewer age suitability (e.g., "13+", "18+")

**runtime:** Duration in minutes

**genres:** One or more genre tags (e.g., "Drama, Thriller")

**production_countries:** Country or countries of origin

**seasons:** Applicable only for shows

**imdb_score, imdb_votes, tmdb_score, tmdb_popularity:** Ratings and popularity metrics from IMDb and TMDb

**Key Findings**

**Type Distribution:** The majority of content is movies, indicating a primary focus on single-session entertainment.

**Age Certification:** Content skewed towards mature audiences, with "16+" and "18+" being the most frequent certifications.

**Genres:** Most common genres are Drama, Comedy, and Action, reflecting general viewer preferences and production trends.

**Release Trends:** Content volume significantly increased after 2015, correlating with Amazon’s investment in original productions.

**Ratings Analysis:** Most titles have IMDb scores between 6 and 7, indicating a fair reception. IMDb and TMDb scores are moderately correlated.

**Runtime Distribution:** Most movies range from 80 to 120 minutes, while most TV shows are limited to 1–2 seasons.

**Production Countries:** USA dominates production, followed by UK, India, and Canada, suggesting regional concentration in content sourcing.

Filling the null values in description with "No description available"

````python
#  description – Only 119 missing
df['description'] = df['description'].fillna("No description available")
````

There were 6484 missing values in age_certification columns , We will fill these empty values with "Unkown"

````python
# age_certification – 6484 missing (categorical)
df['age_certification'] = df['age_certification'].fillna("Unknown")
````

The seasons column has 8511 missing values , so replaceing the empty values with 0.

````python
# seasons – 8511 missing (mostly for MOVIE type)
# Set seasons to 0 for movies, and leave as NaN for shows if needed
df.loc[df['type'] == 'MOVIE', 'seasons'] = df.loc[df['type'] == 'MOVIE', 'seasons'].fillna(0)
````

The imdb_score column has 1021 missing values , we are filling the emplty values with the mean value.

````python
# imdb_score – 1021 missing (important for rating analysis)
df['imdb_score'] = df['imdb_score'].fillna(df['imdb_score'].mean())
````

The imdb_votes column has 1031 missing values , these we will fill by using votes value equal to median.

````python
# imdb_votes – 1031 missing (median)
df['imdb_votes'] = df['imdb_votes'].fillna(df['imdb_votes'].median())
````

The tmdb_popularity column has 547 null values , we will fill these with 0.

````python
# tmdb_popularity – 547 missing (can be filled with 0)
df['tmdb_popularity'] = df['tmdb_popularity'].fillna(0)
````

The tmdb_score has 2080 missing values , these can be filled with mean() values of the same column.

````python
# tmdb_score – 2080 missing (also important for quality analysis)
df['tmdb_score'] = df['tmdb_score'].fillna(df['tmdb_score'].mean())
````

The imdb_id has 667 columns with null values ,since these cannot be filled with any values, beacuse id are unique ,so were are dropping the rows having null values.

````python
df = df.drop(columns=['imdb_id'])
````

**Checking the missing values after doing cleaning of the respective columns.**

````python
print("\nMissing values after cleaning:")
df.isnull().sum()
````

The above output shows that all the null values has been either filled with relevant data or has been dropped.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1  **Distribution of content type**

````python
# Chart - 1 visualization code

# Create count plot
plt.figure(figsize=(4, 5))
ax = sns.countplot(data=df, x='type',color='green')
plt.title('Distribution of Content Type')

# Add value labels
for i in ax.patches:
    count = int(i.get_height())
    ax.annotate(f'{count}',
                (i.get_x() + i.get_width() / 2., i.get_height()),
                ha='center', va='bottom', fontsize=11, color='black')

plt.show()
````

##### 1. Why did you pick the specific chart?

We use countplot because it's ideal for visualizing the frequency of categorical variables. It automatically counts and displays the number of entries in each category, making it easy to compare values like Movies vs Shows in a clear, visual format.



##### 2. What is/are the insight(s) found from the chart?

The count plot depcits that
1. shows are 1357
2. movies are 8511

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights will help create a positive business impact. Identifying that movies (8511) significantly outnumber shows (1357) highlights a content type imbalance. Since TV shows often drive long-term viewer engagement, Amazon Prime can focus on producing or acquiring more shows to improve watch time, customer retention, and subscription value.

**Are there any insights that may lead to negative growth?**

#### Chart - 2 **Distribution of Genres**

````python
# Chart - 2 visualization
plt.figure(figsize=(12, 6))
ax = sns.countplot(
    data=df[df['genres'].isin(df['genres'].value_counts().head(10).index)],
    x='genres',
    hue='genres',
    palette='Set1',
    legend=False
)
plt.title('Top 10 Genres')
plt.xlabel('Genres')
plt.ylabel('Count')

# Add value labels
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

We used a bar chart (countplot) because:

1.The genres column contains categorical values (genre combinations).

2.Bar charts are ideal for showing frequency distribution of categories.

3.It visually highlights the most common genre combinations.



##### 2. What is/are the insight(s) found from the chart?

1.The most frequent genre is [drama] with 908 titles, followed by [comedy] (686 titles) and [documentation] (493 titles).

2.Genre combinations like [drama, romance], [thriller, drama], and [comedy, drama] are also common.

3.There are 209 entries with an empty genre ([]), indicating missing or unclassified genre data.
